### 6. Type of complaint registered more number of times

```python
import nltk
import os
import nltk.corpus
import re
import string
```

```python
Freq_complaints = comcast_consumer['Customer Complaint']
Freq_complaints
```

```python
Freq_complaints = str(Freq_complaints)
Freq_complaints
```

```python
nltk.download('punkt')
corpus = nltk.word_tokenize(Freq_complaints)
```

```python
corpus
```

```python
for i in range(len(corpus )):
    corpus [i] = corpus [i].lower() 
    corpus [i] = re.sub(r'\W',' ',corpus [i]) 
    corpus [i] = re.sub(r'\s+','',corpus [i]) 
```

```python
print(len(corpus))
```

```python
wordfreq = {}
for word in corpus: 
    tokens = nltk.word_tokenize(word) 
    for token in tokens:
        if token not in wordfreq.keys():
            wordfreq[token] = 1 
        else:
            wordfreq[token] += 1
```

```python
print(wordfreq)
```

```python
import heapq
most_freq = heapq.nlargest(20, wordfreq, key=wordfreq.get)
```

```python
most_freq
```

Since the word 'Comcast' is the name of the company and other words such as service, internet and speed are related to the complaint types, from the 'Most_freq' dataset we can derive that more number of complaints registered are regarding the service, internet and speed issue.
